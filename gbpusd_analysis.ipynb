```python
# Function to calculate RSI
def calculate_rsi(data, window=14):
    delta = data.diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()
    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

# Add RSI to the dataset
gbpusd['RSI'] = calculate_rsi(gbpusd['Close'])

# Plot RSI
gbpusd['RSI'].plot(figsize=(12, 6), title="Relative Strength Index (RSI) - GBP/USD")
plt.axhline(70, linestyle="--", color="red", label="Overbought")
plt.axhline(30, linestyle="--", color="green", label="Oversold")
plt.xlabel("Date")
plt.ylabel("RSI")
plt.legend()
plt.grid()
plt.show()
```

```python
# Import necessary libraries
import numpy as np  # For numerical operations
import pandas as pd  # For handling datasets
import matplotlib.pyplot as plt  # For visualizations

# 📌 Load your cleaned GBP/USD dataset (Make sure you've saved it earlier)
gbpusd = pd.read_csv("gbpusd_cleaned.csv", index_col=0, parse_dates=True)

# -----------------------------------------------------------------------------------
# 📌 1. Define Moving Averages for Trend Analysis
# -----------------------------------------------------------------------------------

# Calculate 10-day and 30-day moving averages (MAs)
gbpusd['10-Day MA'] = gbpusd['Close'].rolling(window=10).mean()
gbpusd['30-Day MA'] = gbpusd['Close'].rolling(window=30).mean()

# Generate Buy/Sell signals based on moving average crossover strategy
gbpusd['Signal'] = 0  # Default signal (No trade)
gbpusd.loc[gbpusd['10-Day MA'] > gbpusd['30-Day MA'], 'Signal'] = 1  # Buy Signal
gbpusd.loc[gbpusd['10-Day MA'] < gbpusd['30-Day MA'], 'Signal'] = -1  # Sell Signal

# -----------------------------------------------------------------------------------
# 📌 2. Backtesting: Strategy vs. Market Returns
# -----------------------------------------------------------------------------------

# Calculate daily strategy returns (Shift signal to avoid lookahead bias)
gbpusd['Strategy Return'] = gbpusd['Signal'].shift(1) * gbpusd['Daily Return']

# Plot cumulative returns to compare strategy vs market
gbpusd[['Daily Return', 'Strategy Return']].cumsum().apply(np.exp).plot(figsize=(12, 6), title="Cumulative Returns: Strategy vs. Market")
plt.xlabel("Date")
plt.ylabel("Cumulative Return")
plt.legend(["Market Return", "Strategy Return"])
plt.grid()
plt.show()

# -----------------------------------------------------------------------------------
# 📌 3. Adding RSI (Relative Strength Index) for Better Trade Confirmation
# -----------------------------------------------------------------------------------

def calculate_rsi(data, window=14):
    """
    Function to compute Relative Strength Index (RSI).
    RSI measures market momentum and identifies overbought/oversold conditions.
    """
    delta = data.diff()  # Calculate price changes
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()  # Compute gains
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()  # Compute losses
    rs = gain / loss  # Compute Relative Strength
    rsi = 100 - (100 / (1 + rs))  # Compute RSI
    return rsi

# Add RSI indicator to dataset
gbpusd['RSI'] = calculate_rsi(gbpusd['Close'])

# Plot RSI to visualize overbought/oversold zones
plt.figure(figsize=(12, 6))
plt.plot(gbpusd['RSI'], label="RSI")
plt.axhline(70, linestyle="--", color="red", label="Overbought (Sell)")
plt.axhline(30, linestyle="--", color="green", label="Oversold (Buy)")
plt.title("Relative Strength Index (RSI) - GBP/USD")
plt.xlabel("Date")
plt.ylabel("RSI Value")
plt.legend()
plt.grid()
plt.show()

# -----------------------------------------------------------------------------------
# 📌 4. Risk Management: Stop-Loss and Take-Profit
# -----------------------------------------------------------------------------------

# Define stop-loss and take-profit thresholds (adjust as needed)
stop_loss = -0.02  # Stop loss at -2% loss per trade
take_profit = 0.04  # Take profit at +4% gain per trade

# Apply risk management rules to strategy returns
gbpusd['Risk Adjusted Return'] = gbpusd['Strategy Return'].apply(
    lambda x: stop_loss if x < stop_loss else take_profit if x > take_profit else x
)

# Compare original vs. risk-adjusted returns
gbpusd[['Strategy Return', 'Risk Adjusted Return']].cumsum().apply(np.exp).plot(figsize=(12, 6), title="Strategy Returns With Risk Management")
plt.xlabel("Date")
plt.ylabel("Cumulative Return")
plt.legend(["Original Strategy", "Risk-Managed Strategy"])
plt.grid()
plt.show()

# -----------------------------------------------------------------------------------
# 📌 5. Save Optimized Dataset for Future Use
# -----------------------------------------------------------------------------------

gbpusd.to_csv("gbpusd_final_strategy.csv", index=True)

# -----------------------------------------------------------------------------------
# 📌 6. Summary of Key Takeaways
# -----------------------------------------------------------------------------------

print("✅ Moving Average Crossover Strategy Applied (10-Day vs 30-Day)")
print("✅ RSI Indicator Added for Trade Confirmation")
print("✅ Strategy Performance vs Market Compared")
print("✅ Stop-Loss & Take-Profit Applied for Risk Management")
print("✅ Final Dataset Saved: gbpusd_final_strategy.csv")

```
